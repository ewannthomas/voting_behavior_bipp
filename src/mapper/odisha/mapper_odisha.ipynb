```python
# odisha
df1 = pd.read_excel('India_AC_shapefile.xlsx')
df1.head()
# read sheet 2 of goa masterlist
df2 = pd.read_excel('odisha masterlist.xlsx', sheet_name=0)
# df2 = pd.read_excel('goa_masterlist.xlsx')
df2.rename(columns={'Cleaned': 'cleaned'}, inplace=True)
df2.rename(columns={'assembly constituency': 'AC_NAME'}, inplace=True)
df2.rename(columns={'File status': 'File Status'}, inplace=True)
df2.rename(columns={'Error comment': 'Error Comment'}, inplace=True)
df2.rename(columns={'Year': 'year'}, inplace=True)

df2.head()



def find_fuzzy_match(name, choices):
    if not name:
        return None
    result = process.extractOne(name, choices)
    if result:
        return result[0]
    else:
        return None

# Create a mapping dictionary for rows in df2
ac_name_mapping = {row['AC_NAME']: row for idx, row in df2.iterrows()}

# Filter df1 for rows with ST_NAME "GOA" and map columns from df2 based on fuzzy matching
df1_odisha = df1[df1['ST_NAME'] == 'ODISHA'].copy()  # Filter rows where ST_NAME is "ODISHA"
print(df1_odisha)
df1_odisha['Downloaded'] = df1_odisha['AC_NAME'].apply(lambda x: ac_name_mapping[find_fuzzy_match(x, df2['AC_NAME'])]['Downloaded'])
df1_odisha['year'] = df1_odisha['AC_NAME'].apply(lambda x: ac_name_mapping[find_fuzzy_match(x, df2['AC_NAME'])]['year'])
df1_odisha['Converted'] = df1_odisha['AC_NAME'].apply(lambda x: ac_name_mapping[find_fuzzy_match(x, df2['AC_NAME'])]['Converted'])
df1_odisha['cleaned'] = df1_odisha['AC_NAME'].apply(lambda x: ac_name_mapping[find_fuzzy_match(x, df2['AC_NAME'])]['cleaned'])
df1_odisha['File Status'] = df1_odisha['AC_NAME'].apply(lambda x: ac_name_mapping[find_fuzzy_match(x, df2['AC_NAME'])]['File Status'])
df1_odisha['Error Comment'] = df1_odisha['AC_NAME'].apply(lambda x: ac_name_mapping[find_fuzzy_match(x, df2['AC_NAME'])]['Error Comment'])
df1_odisha['election'] = df1_odisha['AC_NAME'].apply(lambda x: ac_name_mapping[find_fuzzy_match(x, df2['AC_NAME'])]['election'])


df1_odisha.to_excel('Odisha_AC_Mapped_2014.xlsx', index=False)
```

```python
# goa for bye-election 
df1 = pd.read_excel('India_AC_shapefile.xlsx')
df1.head()
# read sheet 2 of goa masterlist
df2 = pd.read_excel('odisha masterlist.xlsx', sheet_name=2)
# df2 = pd.read_excel('goa_masterlist.xlsx')
df2.head()
print(df2)


def find_fuzzy_match(name, choices):
    if not name:
        return None
    result = process.extractOne(name, choices)
    if result:
        return result[0]
    else:
        return None

# Create a mapping dictionary for rows in df2
ac_name_mapping = {row['AC_NAME']: row for idx, row in df1.iterrows()}

# Filter df1 for rows with ST_NAME "GOA" and map columns from df2 based on fuzzy matching
df1_odisha = df1[df1['ST_NAME'] == 'ODISHA'].copy()  # Filter rows where ST_NAME is "GOA"
print(df1_odisha)




df2['FID'] = df2['AC_NAME'].apply(lambda x: ac_name_mapping[find_fuzzy_match(x, df1_odisha['AC_NAME'])]['FID'])
df2['OBJECTID'] = df2['AC_NAME'].apply(lambda x: ac_name_mapping[find_fuzzy_match(x, df1_odisha['AC_NAME'])]['OBJECTID'])
df2['ST_CODE'] = df2['AC_NAME'].apply(lambda x: ac_name_mapping[find_fuzzy_match(x, df1_odisha['AC_NAME'])]['ST_CODE'])
df2['ST_NAME'] = df2['AC_NAME'].apply(lambda x: ac_name_mapping[find_fuzzy_match(x, df1_odisha['AC_NAME'])]['ST_NAME'])
df2['DT_CODE'] = df2['AC_NAME'].apply(lambda x: ac_name_mapping[find_fuzzy_match(x, df1_odisha['AC_NAME'])]['DT_CODE'])
df2['DIST_NAME'] = df2['AC_NAME'].apply(lambda x: ac_name_mapping[find_fuzzy_match(x, df1_odisha['AC_NAME'])]['DIST_NAME'])
df2['AC_NO'] = df2['AC_NAME'].apply(lambda x: ac_name_mapping[find_fuzzy_match(x, df1_odisha['AC_NAME'])]['AC_NO'])
df2['PC_NO'] = df2['AC_NAME'].apply(lambda x: ac_name_mapping[find_fuzzy_match(x, df1_odisha['AC_NAME'])]['PC_NO'])
df2['PC_NAME'] = df2['AC_NAME'].apply(lambda x: ac_name_mapping[find_fuzzy_match(x, df1_odisha['AC_NAME'])]['PC_NAME'])
df2['PC_ID'] = df2['AC_NAME'].apply(lambda x: ac_name_mapping[find_fuzzy_match(x, df1_odisha['AC_NAME'])]['PC_ID'])
df2['STATUS'] = df2['AC_NAME'].apply(lambda x: ac_name_mapping[find_fuzzy_match(x, df1_odisha['AC_NAME'])]['STATUS'])
df2['Shape_Leng'] = df2['AC_NAME'].apply(lambda x: ac_name_mapping[find_fuzzy_match(x, df1_odisha['AC_NAME'])]['Shape_Leng'])
df2['Shape_Area'] = df2['AC_NAME'].apply(lambda x: ac_name_mapping[find_fuzzy_match(x, df1_odisha['AC_NAME'])]['Shape_Area'])

# drop AC no
df2.drop(['AC No.'], axis=1, inplace=True)
# print(df1)
df2.to_excel('Odisha_AC_Mapped_bye.xlsx', index=False)
```
